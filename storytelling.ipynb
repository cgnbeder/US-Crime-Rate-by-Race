On the Misinterpretation of Racism and the Role of the Media:

I believe that racism is often misunderstood in today's context. Racism and baseless hatred have become conflated, causing both terms to lose their true meanings. Racism, in essence, is an instinctual perception driven by the human body's survival mechanism, influenced by factors such as media, news, observations, and personal experiences. Our brains use visual cues to categorize negative events or situations. As a result, racial characteristics, which constitute a significant part of a person's appearance, become the most noticeable distinguishing visual features.

For example, in a society composed solely of white people, if a group of people wearing orange clothing attacks you in the street for no apparent reason, you might develop an unconscious survival instinct against people wearing orange. The next time you walk down the street, you might be more cautious around those in orange clothing. This is not an irrational hatred towards the color orange but rather a natural defensive mechanism triggered by your body in response to a negative experience.

Similarly, there is a form of "racism" that can be perceived as positive. Generalizations like "Japanese people are hardworking and punctual," "Germans are disciplined," or "French people are romantic" are also formed through media, news, observations, and experiences, where the brain uses race as a broad category to make sense of the world. In this sense, racism is not always negative.

Therefore, the instincts you develop based on the news you watch, the media you follow, and your personal experiences may make you "racist" in a sense, but this does not make you a bad person unless you wake up one day with a baseless hatred for a group or race. Hatred and racism are distinct concepts and should not be confused.

I am undertaking this project out of personal curiosity and a desire to provide a more impartial, third-party perspective on how the media uses crime and racial distribution rates to serve their respective beliefs. On one side, certain media outlets conclude their analyses with purely racist or hateful interpretations, relying solely on statistics about crime rates and racial differences without considering third-party factors. On the other side, anti-racist and humanist groups and media, in stark contrast, ignore these numbers entirely and base their conclusions solely on third-party factors. However, numbers provide us with insights, not conclusions. The goal is to use scientific data to gain insights and align these qualitative studies with third-party factors impartially, which is the essence of analytical thinking.
